```python
import pandas as pd
url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_2021-01.csv.gz'
df1 = pd.read_csv(url,nrows =50)
print(df1.dtypes)
taxi_dtypes ={
    'VendorID': pd.Int64Dtype(),
    'passenger_count': pd.Int64Dtype(),
    'trip_distance': float,
    'RatecodeID':pd.Int64Dtype(),
    'store_and_fwd_flag':str,
    'PULocationID':pd.Int64Dtype(),
    'DOLocationID':pd.Int64Dtype(),
    'payment_type': pd.Int64Dtype(),
    'fare_amount': float,
    'extra':float,
    'mta_tax':float,
    'tip_amount':float,
    'tolls_amount':float,
    'improvement_surcharge':float,
    'total_amount':float,
    'congestion_surcharge':float}
```

```python
parse_dates = ['tpep_pickup_datetime', 'tpep_dropoff_datetime']
type(parse_dates)
df = pd.read_csv(url, compression='gzip', dtype=taxi_dtypes, parse_dates=parse_dates, nrows=1000)

df.dtypes

```

```python
passenger0 = df['passenger_count'].isin([0]).sum()
print(f"No of rows before  {len(df)}")
print(f"Sum of rows data with 0 passenger is {passenger0}")

df = df[df['passenger_count']>0]
passenger0 = df['passenger_count'].isin([0]).sum()
print(f"Sum of rows data with 0 passenger after transform is {passenger0}")
print(f"No of rows after  {len(df)}")
```
